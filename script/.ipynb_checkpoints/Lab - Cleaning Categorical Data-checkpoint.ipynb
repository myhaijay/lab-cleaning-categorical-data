```python
#Create a function to plot every discrete variables. Do the same with continuous variables 
#(be careful, you may change the plot type to another one better suited for continuous data).

def plot_discrete_vrbs(df):
    thresh_hold = 100
    discrete_cols = [column for column in df.columns if df[column].nunique() <= thresh_hold]
    for col in discrete_cols:
        discrete_plots = sns.displot(df[col]) 
    return discrete_plots
```

```python
#continuous_cols = [column for column in df.columns if df[column].nunique() >= thresh_hold]
def plot_continuous_vrbs(df):
    thresh_hold = 100
    continuous_cols = [column for column in df.columns if df[column].nunique() >= thresh_hold]
    for col in continuous_cols:
        plt.figure()
        continuous_plots = sns.distplot(df[col], hist=False) #can remove the hist arguement to see the histograms 
    return continuous_plots
```

7. Have you found outliers? If you have, what should we do with them?
We see outliers in 3 of the continuous variables. 
It's advisible to remove the outliers so as to get better results overall.

```python
#Removing the outliers
def remove_outliers(df, threshold=1.5, in_columns=['Customer Lifetime Value', 'Monthly Premium Auto', 'Total Claim Amount'],
                    skip_columns=[]):
    for column in in_columns:
        if column not in skip_columns:
            upper = np.percentile(df[column],75)
            lower = np.percentile(df[column],25)
            iqr = upper - lower
            upper_limit = upper + threshold * iqr
            lower_limit = lower - threshold * iqr
            df = df[(df[column]>lower_limit) & (df[column]<upper_limit)]
    return df
```

```python
continuous_df_no_outlier = remove_outliers(continuous_df)

for col in continuous_df_no_outlier.columns:
    sns.boxplot(continuous_df_no_outlier[col])
    plt.show()
```

```python
# 9. Define a function that differentiate between continuous and discrete variables. 
#Hint: Number of unique values might be useful. Store continuous data into a continuous variable and do the same for discrete and categorical.
def differentiate_variables(df):

    continuous_columns = []
    discrete_columns = []
    thresh_hold = 100

    for i in df.columns:
        if df[i].nunique() >= thresh_hold:
            continuous_columns.append(i)
        else:
            discrete_columns.append(i)
    continuous_variables = df[continuous_columns]
    discrete_variables = df[discrete_columns]
    return discrete_variables, continuous_variables
```

```python
discrete, continuous   = differentiate_variables(data)
discrete
```
